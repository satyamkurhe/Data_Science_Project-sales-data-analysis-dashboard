## üì¶ Step 1: Import Required Libraries

## üìÅ Step 2: Load Dataset
Dataset columns:
- Order ID
- Date
- Product
- Category
- Quantity
- Price
- Region


```python
df = pd.read_csv("sales_data.csv")
df.head()
```

---

## üìä Step 3: Monthly Sales Trend


```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Price'].sum()

monthly_sales.plot(kind='bar', color='skyblue', figsize=(10,5))
plt.title(" Monthly Revenue")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

---

## üèÜ Step 4: Top-Selling Products


```python
top_products = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)

top_products.plot(kind='bar', color='green')
plt.title(" Top 5 Products by Quantity Sold")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
